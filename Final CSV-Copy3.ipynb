```python
Other_Cancer_dummy = pd.get_dummies(df.Other_Cancer,prefix = 'Other_C', drop_first = True)
Other_Cancer_dummy.head()
```

```python
Diabetes_dummy = pd.get_dummies(df.Diabetes, prefix = 'Diabetes', drop_first = True)
Diabetes_dummy.head()
```

```python
Diabetes_dummy = Diabetes_dummy.rename(columns = {'Diabetes_No, pre-diabetes or borderline diabetes': '!Diabetes_pre_bor', 'Diabetes_Yes, but female told only during pregnancy': 'Diabetes_Yes_Pregnancy'})
Diabetes_dummy.head()
```

```python
Arthritis_dummy = pd.get_dummies(df.Arthritis,prefix = 'Arthititis', drop_first = True)
Arthritis_dummy.head()
```

```python
Sex_dummy = pd.get_dummies(df.Sex,prefix = 'Sex', drop_first = True)
Sex_dummy.head()
```

```python
Age_dummy = pd.get_dummies(df.Age_Category,prefix = 'Age', drop_first = True)
Age_dummy.head()
```

```python
Smoking_dummy = pd.get_dummies(df.Smoking_History, prefix ='Smoking', drop_first = True)
Smoking_dummy.head()
```

concating the dummies with original dataframe

```python
final = pd.concat([df,General_dummy,Checkup_dummy, Exercise_dummy, Heart_Disease_dummy, Skin_Cancer_dummy, Other_Cancer_dummy, Depression_dummy, Diabetes_dummy, Arthritis_dummy, Sex_dummy, Age_dummy, Smoking_dummy], axis = 1)
final.head()
```
